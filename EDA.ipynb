# Lambda

```python
import pandas as pd
```

```python
df = pd.read_excel('Result.xlsx',index_col = 'Roll')
```

```python
df
```

```python
df.columns
```

```python
df['Total']=df.Python + df.Pandas +df.Numpy + df.Stats + df.Flask + df.ML
```

```python
df
```

```python
df['Percentage'] = round((df.Total/600)*100,2)
```

```python
df
```

```python
def grades(percentage):
    if percentage >=90:
        return 'A+'
    elif percentage >=80:
        return 'A'
    elif percentage >=70:
        return 'B'
    elif percentage >=60:
        return 'C'
    elif percentage>=50:
        return 'D'
    else:
        return 'F'
```

```python
df['Grade'] = df.Percentage.apply(grades)
```

```python
df
```

```python
df['Grade1'] = df.Percentage.apply(lambda x: "A+" if x>=90 else "A" if x>80 else "B" if x>=70 else "C" if x>=60 else "D" if x>=50 else "Fail")
```

```python
df
```

```python
df['Grade2'] = ["A+" if x>=90 else "A" if x>=80 else "B" if x>=70 else "C" if x>=60 else "D" if x >=50 else "Fail" for x in df.Percentage]
```

```python
df
```

```python
def status(marks):
    Py,Pd,Np,St,Fl,Ml = marks
    if Py >= 60:
        if Pd >=60:
            if Np >=60:
                if St >=60:
                    if Fl >= 60:
                        if Ml >=60:
                            return "Pass"
                        else:
                            return "Fail"
                    else:
                        return "Fail"
                else:
                    return "Fail"
            else:
                return "Fail"
        else:
            return "Fail"
    else:
        return "Fail"
```

```python
df['Status'] = [status(a) for a in zip(df.Python,df.Pandas,df.Numpy,df.Stats,df.Flask,df.ML)]
```
